```python
#Ans1:
import seaborn as sns
import plotly.express as px
titanic = sns.load_dataset('titanic')
fig = px.scatter(titanic, x='age', y='fare', title='Scatter Plot of Age vs. Fare')

fig.show()
```

```python
#Ans2:
import plotly.express as px
tips = px.data.tips()
fig = px.box(tips, x='day', y='total_bill', title='Box Plot of Total Bill by Day')
fig.show()
```

```python
#Ans3:
tips = px.data.tips()
fig = px.histogram(
    tips,
    x='sex',   
    y='total_bill',     
    color='day',      
    pattern_shape='smoker', 
    title='Histogram of Total Bill by Sex with Smoker and Day Information'
)
fig.show()
```

```python
#Ans4:

iris = px.data.iris()
fig = px.scatter_matrix(
    iris,
    dimensions=['sepal_length', 'sepal_width', 'petal_length', 'petal_width'], 
    color='species', 
    title='Scatter Matrix of Iris Dataset'
)
fig.show()

```

```python
#Ans5: Distplot is a function from the seaborn library, primarily used for visualizing the distribution of a single variable.

iris = px.data.iris()
fig = px.histogram(
    iris,
    x='sepal_length',       
    color='species',          
    marginal='rug',  
    title='Distribution of Sepal Length with KDE'
)
fig.show()
```
